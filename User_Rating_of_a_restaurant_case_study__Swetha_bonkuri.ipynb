```python
# Feature 1: Number of Votes Given to the Restaurant
# 1. Number of Votes Given to Restaurant:
# Import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd


votes = df['Votes']
rating = df['Aggregate rating']

# Create a figure for the plot with a specific size
plt.figure(figsize=(10, 6))

# Create a scatter plot with 'Votes' on the x-axis and 'Rating' on the y-axis
plt.scatter(votes, rating, alpha=0.6, color='skyblue', edgecolors='navy', linewidth=0.5)

# Set labels for the x-axis and y-axis
plt.xlabel("Number of Votes")
plt.ylabel("Rating")

# Set the title of the plot
plt.title("Relationship between Number of Votes and Rating")

# Add grid lines to the plot with a dashed style and reduced opacity
plt.grid(True, linestyle='--', alpha=0.6)

# Customize tick labels for both axes
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Display the plot
plt.show()
```

```python
# 2. Restaurant Serving More Number of Cuisines:
# Import necessary libraries
# Import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd



# Check for NaN values in the 'Cuisines' column and replace them with an empty string
df['Cuisines'].fillna('', inplace=True)

# Split the non-empty 'Cuisines' values and calculate the number of cuisines
df['Num Cuisines'] = df['Cuisines'].apply(lambda x: len(x.split(',')) if x else 0)
num_cuisines = df['Num Cuisines']

# Create a figure for the plot with a specific size
plt.figure(figsize=(10, 6))

# Create a scatter plot with 'Num Cuisines' on the x-axis and 'Rating' on the y-axis
plt.scatter(num_cuisines, rating, alpha=0.6, color='lightcoral', edgecolors='maroon', linewidth=0.5)

# Set labels for the x-axis and y-axis
plt.xlabel("Number of Cuisines")
plt.ylabel("Rating")

# Set the title of the plot
plt.title("Relationship between Number of Cuisines Served and Rating")

# Add grid lines to the plot with a dashed style and reduced opacity
plt.grid(True, linestyle='--', alpha=0.6)

# Customize tick labels for both axes
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Display the plot
plt.show()

```

```python
# Cuisines Served (Pie Chart):
# A pie chart is a good choice to visualize the division of restaurants based on cuisines served.
# This is because a pie chart shows the relative sizes of different parts of a whole,
# which is helpful in understanding the proportion of restaurants that serve each cuisine.
import matplotlib.pyplot as plt
import pandas as pd

# Extract the cuisines data
cuisines = df["Cuisines"].str.split(', ').explode()

# Count the occurrences of each cuisine
cuisine_counts = cuisines.value_counts()

# Create a figure for the pie chart with a specific size
plt.figure(figsize=(8, 8))

# Create a pie chart to visualize the division of restaurants based on cuisines
plt.pie(cuisine_counts, labels=cuisine_counts.index, autopct="%1.1f%%", startangle=140)

# Set the title of the plot
plt.title("Division of Restaurants Based on Cuisines")

# Display the plot
plt.show()


```

```python
# Locality (Bar Graph):
# A bar graph is a good choice to visualize the division of restaurants based on locality.
# This is because a bar graph shows the number of restaurants in each locality,
# which is helpful in understanding the distribution of restaurants across different localities.


# Extract the locality data
locality = df["Locality"].value_counts().index.to_list()

# Create a figure for the bar graph with a specific size
plt.figure(figsize=(12, 6))

# Create a bar graph to visualize the number of restaurants in each locality
plt.bar(locality, df["Locality"].value_counts().values)

# Set labels for the x-axis and y-axis
plt.xlabel("Locality")
plt.ylabel("Number of Restaurants")

# Set the title of the plot
plt.title("Division of Restaurants Based on Locality")

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

```python
# Price Range (Histogram):
# A histogram is a good choice to visualize the division of restaurants based on price range.
# This is because a histogram shows the distribution of data in a continuous variable,
# which is helpful in understanding the range of prices offered by restaurants.


# Extract the price range data
price_range = df["Price range"]

# Create a figure for the histogram with a specific size
plt.figure(figsize=(10, 6))

# Create a histogram to visualize the division of restaurants based on price range
plt.hist(price_range, bins=range(1, 7), rwidth=0.8, align='left')

# Set labels for the x-axis and y-axis
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")

# Set the title of the plot
plt.title("Division of Restaurants Based on Price Range")

# Set the x-axis ticks to match the price range values
plt.xticks(range(1, 6))

# Display the plot
plt.show()
```
